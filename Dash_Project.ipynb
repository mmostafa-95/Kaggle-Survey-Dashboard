```python
fig3 = dbc.Col(
    html.Div(
        [
            #GRAPH 3    
            dcc.Dropdown(
                id = "my_dropdown2" ,
                options=[  
                    {"label" : str(c) ,"value" : str(c) } for  c in langlist ],
                value = "value" ,
                placeholder = "Choose a Language ... " , 
                multi = True ,
                style = {"color" : "black"}
            ),
            html.Button(id = "button2" , n_clicks = 0 , children = "Submit" , style = {"textAlign" : "center"}),
            html.Br(),
            dcc.Graph( id = "my_fig2") 
                ]
    )
)
```

```python
fig4 = dbc.Col(
    html.Div(
        [
            #GRAPH 4    
            dcc.Dropdown(
                id = "my_dropdown3" ,
                options=[  
                    {'label': k, 'value': k} for k in all_options.keys()],
                value = "value" ,
                placeholder = "Choose an Option ... " , 
                multi = False ,
                style = {"color" : "black"}
            ),
            dcc.Dropdown(
                id = "my_dropdown4" ,
                value = "value" ,
                placeholder = "Choose ... " , 
                multi = True ,
                style = {"color" : "black"}
            ),
            html.Button(id = "button3" , n_clicks = 0 , children = "Submit" , style = {"textAlign" : "center"}),
            dcc.Graph( id = "my_fig3") 
        ]
    )
)
```

```python
app = dash.Dash(external_stylesheets=[dbc.themes.QUARTZ])
```

```python
app.layout = html.Div([
    dbc.Row(dbc.Col(html.H1('Kaggle Survey Dashboard!'), width={'size':6, 'offset':4})),
    dbc.Row(dbc.Col(html.Br())),
    dbc.Row(dbc.Col(html.Br())),
    dbc.Row([first_jumbotron, second_jumbotron, third_jumbotron, fourth_jumbotron]),
    dbc.Row(dbc.Col(html.Br())),
    dbc.Row(dbc.Col(html.H2('Countries of Participants'), width={'size':6, 'offset':4})),
    dbc.Row(dbc.Col(html.Br())),
    dbc.Row(dbc.Col(dcc.Graph(id='Treemap', figure=fig))),
    dbc.Row([fig2, fig3]),
    dbc.Row(fig4)
])
```

```python
@app.callback(
    Output(component_id = "my_fig" , component_property = "figure") , 
    State(component_id = "my_dropdown" , component_property = "value"),
    State(component_id = "my_dropdown1" , component_property = "value"),
    State(component_id = "radio" , component_property ="value"),
    Input(component_id = "button" , component_property = "n_clicks")  
)
def update_1(lang , lang2, radio  , n_clicks ) :
    if radio == "Pie" :
        fig1 = px.pie(data_frame = df , names = lang , hole = 0.4 ,  template = "plotly_dark")
        return fig1
    else :
        fig1 = px.histogram(data_frame = df , x = lang , color = lang2  , barmode = "group" , animation_frame = "Country" , template = "plotly_dark" , title= f"{lang} Grouped by {lang2}")
        return fig1
    
@app.callback(
    Output(component_id = "my_fig2" , component_property = "figure") ,
    State(component_id = "my_dropdown2" , component_property = "value"),
    Input(component_id = "button2" , component_property = "n_clicks")
)
def update_2(lang , n_clicks):
    fig2 = px.histogram(data_frame = df , x = lang ,  template = "plotly_dark" ,title= f"{lang}")
    return fig2

@app.callback(
    Output('my_dropdown4', 'options'),
    [Input('my_dropdown3', 'value')])
def set_options(selected_option):
    return [{'label': i, 'value': i} for i in all_options[selected_option]]

@app.callback(
    Output(component_id = "my_fig3" , component_property = "figure") ,
    State(component_id = "my_dropdown4" , component_property = "value"),
    Input(component_id = "button3" , component_property = "n_clicks")
)
def update_4(lang , n_clicks):
    fig3 = px.histogram(data_frame = df , x = lang  , template = "plotly_dark", title= f"{lang}" )
    return fig3
```

```python
app.run_server()
```
